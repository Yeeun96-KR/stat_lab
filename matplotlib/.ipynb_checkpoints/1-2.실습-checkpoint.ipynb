# 한글깨짐방지

```python
import warnings
warnings.filterwarnings('ignore')

# '-', 한글 깨짐 해결 모듈
import matplotlib.pyplot as plt
# '-' 깨짐 해결 모듈
import matplotlib as mpl

plt.rcParams["font.family"] = "Malgun Gothic"  # For Windows
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (8,4)
print(plt.rcParams["font.family"])

# 마이너스 깨짐 해결
mpl.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start='2021-01-01', end='2021-12-31', freq='M')

np.random.seed(0)
product_a_sales = np.random.normal(loc=100, scale=20, size=len(dates))
product_b_sales = np.random.normal(loc=150, scale=30, size=len(dates))
product_c_sales = np.random.normal(loc=80, scale=15, size=len(dates))

sales_data = pd.DataFrame({
    'Date':dates,
    'Product A':product_a_sales,
    'Product B':product_b_sales,
    'Product C':product_c_sales
})

print(sales_data.head())
```

```python
product_a_saels = sales_data['Product A']

plt.hist(product_a_sales, bins=14, edgecolor='black')

plt.title('Histogram of Product A Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')

plt.show()
```

```python
iris_data = pd.read_csv('datas/iris.csv')

sepal_length = iris_data['sepal_length']

plt.hist(sepal_length, bins=20, edgecolor='black')

plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

plt.show()
```

```python
fruits = ['Apple', 'Banana', 'Orange', 'Grape', 'Watermelon']
sales = [30, 40, 25, 15, 20]

plt.pie(sales, labels=fruits, autopct='%1.1f%%', startangle=90)

plt.title('Sales of Fruits')

plt.show()
```

```python
wedgeprops={'width':0.6, 'edgecolor':'w', 'linewidth':3}

plt.pie(sales, labels=fruits, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)

plt.title('Sales of Fruits')

plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C']
ratios = [30, 40, 20]

plt.pie(ratios, labels=categories, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor':'white'})
center_circle = plt.Circle((0,0), 0.5, color='White', edgecolor='white')

fig=plt.gcf()
fig.gca().add_artist(center_circle)

plt.title('Category Ratios')

plt.show()
```

```python
data = {
    'Height': [165, 170, 175, 180, 185],
    'Weight': [60, 70, 80, 85, 90]
}

df = pd.DataFrame(data)

plt.scatter(df['Height'], df['Weight'])

plt.xlabel('Height')
plt.ylabel('Weight')

plt.xlim(164, 187)
plt.show()
```

```python
height = [175,170,162,171,186,175,166,169,189,178]
weight = [70,65,61,63,78,64,60,64,88,76]

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'I', 'j']
plt.scatter(height, weight)
```

```python
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'I', 'j']
plt.scatter(height, weight)

for i, j, k in zip(labels, height, weight):
    plt.annotate(i,
                 xy=(j,k),
                 xytext=(5,-5),
                 textcoords='offset points')

plt.title('height vs.weight')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
import seaborn as sns
```

```python
data = {
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05'],
    'Stock1': [100, 110, 105, 95, 120],
    'Stock2': [80, 85, 90, 95, 100],
    'Stock3': [120, 130, 125, 135, 140],
    'Stock4': [70, 75, 80, 85, 90]
}

df = pd.DataFrame(data)
df

df.set_index('Date', inplace=True)

sns.heatmap(df, cmap='magma', annot=True, fmt='g')

plt.title('Stock Prices Heatmap')

plt.show()
```

```python
df_cor = df.corr()

sns.heatmap(df_cor, cmap='viridis', annot=True, fmt='g')

plt.title('Stock Prices heatmap')

plt.show()
```
