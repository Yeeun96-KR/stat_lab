```python
file_path = 'datas/sns_tips.csv'
tips = pd.read_csv(file_path)
tips.head()
```

```python
tips.select_dtypes(include='object').columns
```

```python
tips['gender'].value_counts().sort_index()
```

```python
gen_map = {
    'Female':0,
    'Male':1
}
```

```python
tips['smoker'].value_counts().sort_index()
```

```python
smo_map={
    'No':0,
    'Yes':1
}
```

```python
tips['day'].value_counts().sort_index()
```

```python
day_map={
    'Fri':0,
    'Thur':1,
    'Sat':2,
    'Sun':3
}
```

```python
tips['time'].value_counts().sort_index()
```

```python
time_map={
    'Dinner':0,
    'Lunch':1
}
```

```python
tips['gender'] = tips['gender'].map(gen_map)
tips['smoker'] = tips['smoker'].map(smo_map)
tips['day'] = tips['day'].map(day_map)
tips['time'] = tips['time'].map(time_map)
```

```python
plt.title('heatmap')
sns.heatmap(tips.corr(), annot=True, linewidths=1)
plt.show()
```

```python
plt.title('heatmap')
sns.heatmap(tips.corr(),
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            linewidths=0.5,
            linecolor='gray',
            cbar=True,
            vmin=-1,
            vmax=1
           )
plt.show()
```
