```python
columns = ['survived', 'pclass', 'name', 'sex', 'age', 'sibsp', 'parch', 'fare']
t_df1 = t_df[columns].copy()
```

```python
tmp = []
for row in t_df1['sex']:
    if row == 'female':
        tmp.append(0)
    else:
        tmp.append(1)

tmp
```

```python
t_df1['gender'] = tmp
```

```python
t_df1.loc[:,'gender'] = tmp
```

```python
t_df1
```

```python
t_df1.drop(columns='sex', inplace=True)
t_df1.head()
```

```python
t_df1.info()
```

```python
t_df1['name']
```

```python
condition = lambda x : x.split(',')[1].split('.')[0].strip()

t_df1.loc[:,'title'] = t_df1['name'].apply(condition)
# t_df1.loc[:,'title'] = t_df1['name'].map(condition)
t_df1['title'].unique()
```

```python
specials = ['Master', 'Don', 'Rev']

for special in specials:
    t_df1['title'] = t_df1['title'].replace(special, 'Special')

t_df1['title'].unique()
```

```python
len(t_df1['title'].unique())
```

```python
t_df1 = t_df1.drop('name', axis=1)
t_df1
```

```python
def convert_title(x):
    if x == 'Special':
        return 1
    else:
        return 0

t_df1['special_title'] = t_df1['title'].map(convert_title)
t_df1.tail()
```

```python
convert_title = lambda x : 1 if x == 'Special' else 0

t_df1['special_title'] = t_df1['title'].apply(convert_title)
t_df1.tail()
```

```python
t_df1.drop('title', axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pclass_skew = t_df['pclass'].skew()
plt.figure(figsize=(8,6))
sns.histplot(t_df['pclass'], kde=True, bins=10)
plt.title(f'Distribution of Pclass with Skewness: {pclass_skew:2f}')
plt.xlabel('Pclass')
plt.ylabel('Frequency')
plt.show()
```

```python
fare_skew = t_df['fare'].skew()
plt.figure(figsize=(8,6))
sns.histplot(t_df['fare'], kde=True, bins=30)
plt.title(f'Distribution of Fare with Skewness: {fare_skew:2f}')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.show()
```

```python
numeric_kurt = t_df.select_dtypes(include='number').kurt()
numeric_kurt
```

```python
fare_kurtosis = t_df['fare'].kurt()
plt.figure(figsize=(8,6))
sns.histplot(t_df['fare'], kde=True, bins=30)

plt.title(f'Distribution of Fare with kurtosis : {fare_kurtosis:.2f}')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
