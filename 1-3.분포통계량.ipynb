```python
x = np.arange(1,12,1)
print(x)
```

### 왜도 계산

```python
from matplotlib import pyplot as plt
```

```python
x1 = [1] * 30 + [2] * 20 + [3] * 20 + [4] * 15 + [5] * 15 # 좌측으로 치우침
x2 = [1] * 15 + [2] * 20 + [3] * 30 + [4] * 20 + [5] * 15 # 치우치지 않음
x3 = [1] * 15 + [2] * 15 + [3] * 20 + [4] * 20 + [5] * 30 # 우측으로 치우침
```

```python
print(x1)
```

```python
from collections import Counter
print(Counter(x1))
```

```python
print(Counter(x2))
```

```python
print(Counter(x3))
```

```python
pd.Series(x1).plot(kind='hist', bins=8)
```

```python
pd.Series(x1).plot(kind='kde') 
# kernel density estimation : distribution of ovservations in a dataset
```

```python
pd.Series(x1).value_counts(sort=False) 
# 빈도수를 계산하되 결과를 정렬하지 않고 원래 데이터 순서대로 반환
```

```python
pd.Series(x1).value_counts(sort=False).plot(kind='bar')
```

```python
pd.Series(x2).value_counts()
```

```python
pd.Series(x2).plot(kind='kde')
```

```python
print("좌로 치우쳤을 때 왜도:", skew(x1))
print("치우치지 않았을 때 왜도:", skew(x2))
print("우로 치우친 왜도:", skew(x3))
```

### 첨도 계산

```python
# 첨도 계산을 위한 데이터 생성
x1 = [1] * 20 + [2] * 20 + [3] * 20 + [4] * 20 + [5] * 20 # 전혀 뾰족하지 않음
x2 = [1] * 10 + [2] * 20 + [3] * 40 + [4] * 20 + [5] * 10 # 조금 뾰족
x3 = [1] * 5 + [2] * 15 + [3] * 60 + [4] * 15 + [5] * 5 # 매우 뾰족
```

```python
print("전혀 뾰족하지 않을 때의 첨도 : ", kurtosis(x1))
print("조금 뾰족할 때의 첨도 : ", kurtosis(x2))
print("매우 뾰족할 때의 첨도 : ", kurtosis(x3))
```

```python
pd.Series(x1).value_counts(sort = False).plot(kind = 'bar')
```

```python
pd.Series(x2).value_counts(sort = False).plot(kind = 'bar')
```

```python
pd.Series(x3).value_counts(sort = False).plot(kind = 'bar')
```
